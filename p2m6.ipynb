```python
import cv2
import os
import matplotlib.pyplot as plt
import glob

# Baixar o arquivo Haar Cascade
!wget -O haarcascade_frontalface_default.xml https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml

# Carregar o classificador pré-treinado
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Ler a imagem
video = cv2.VideoCapture('./la_cabra.mp4')


frameno = 0
while(True):
   ret,frame = video.read()
   if ret:
      # se conseguiu pegar um frame:

      gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # converte pra escala de cinza
      faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=10, minSize=(30, 30))

      for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

      # Converter a imagem de BGR para RGB para exibição com Matplotlib
      #img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

      name = str(frameno) + '.jpg'
      print ('new frame captured...' + name)

      cv2.imwrite(name, frame)
      frameno += 1

      # plt.imshow(img_rgb)
      # plt.axis('off')  # Ocultar os eixos
      # plt.show()

   else:
      break


img_array = []
for filename in glob.glob('*.jpg'):
   img = cv2.imread(filename)
   height, width, layers = img.shape
   size = (width,height)
   img_array.append(img)

   out = cv2.VideoWriter('video.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)

for i in range(len(img_array)):
   out.write(img_array[i])
out.release()
```
